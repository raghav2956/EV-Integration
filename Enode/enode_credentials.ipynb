## Get Access Token

```python
# Replace these with your actual client ID and client secret
client_id = 'fd9214df-119c-4444-8f77-5138b9c0d44a'
client_secret = 'f92a87b2fd96fdc96e4ea97a7b45af9923b0195d'

#https://enode-api.production.enode.io/

# Define the URL and the payload
url = 'https://oauth.production.enode.io/oauth2/token'
payload = {
    'grant_type': 'client_credentials'
}

# Make the POST request with basic authentication
response = requests.post(url, auth=HTTPBasicAuth(client_id, client_secret), data=payload)

# Check if the request was successful
if response.status_code == 200:
    token_data = response.json()
    print(token_data)
    access_token = token_data['access_token']
    print(f"Access Token: {access_token}")
else:
    # Print an error message if the request failed
    print(f"Failed to retrieve token: {response.status_code}, {response.text}")
```
